```python
sns.boxplot(x='day', y='total_bill', hue='sex', data=tips_df)
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
```

# Strip Plot

```python
plt.figure(figsize=(8, 5))

sns.stripplot(
  x='day', 
  y='total_bill',
  data=tips_df, 
  jitter=True, # làm gọn các chấm
  hue='sex', # vẽ màu theo giới tính
  dodge=True # tách 2 nhóm ra
)
```

```python
flights = sns.load_dataset('flights')
flights = flights.pivot_table(index='month', columns='year', values='passengers')
flights
```

```python
sns.heatmap(flights, cmap='Blues', linecolor='white', linewidths=1)
ax: axes.Axes = plt.gca()
ax.yaxis.set_tick_params(labelrotation=0)
```

# Pair Grid

```python
iris_g = sns.PairGrid(iris, hue='species')
iris_g.map(sns.scatterplot)
```

```python
iris_g = sns.PairGrid(iris, hue='species')
iris_g.map_diag(sns.histplot)
iris_g.map_offdiag(sns.scatterplot)
```

```python
iris_g = sns.PairGrid(iris, hue='species')
iris_g.map_diag(sns.histplot)
iris_g.map_upper(sns.scatterplot)
iris_g.map_lower(sns.kdeplot)
```

```python
iris_g = sns.PairGrid(iris, hue='species', 
                      x_vars=['sepal_length', 'sepal_width'],
											y_vars=['petal_length', 'petal_width'])

iris_g.map(sns.scatterplot)
iris_g.add_legend()
```

# Facet Grid

```python
tips_fg = sns.FacetGrid(tips_df, col='time', row='smoker')
tips_fg.map(sns.histplot, 'total_bill', bins=10)
```

```python
tips_fg = sns.FacetGrid(tips_df, col='time', row='smoker')
tips_fg.map(sns.scatterplot, 'total_bill', 'tip')
```

```python
tips_fg = sns.FacetGrid(tips_df, col='time', hue='smoker', height=4, aspect=1.3)
tips_fg.map(sns.scatterplot, 'total_bill', 'tip')
```

```python
tips_fg = sns.FacetGrid(tips_df, col='time', hue='smoker', height=4, aspect=1.3,
                        col_order=['Dinner', 'Lunch'], palette='Set1')
tips_fg.map(sns.scatterplot, 'total_bill', 'tip', edgecolor='w')
```

```python
kws = dict(s=50, lw=.5, edgecolor='w')
tips_fg = sns.FacetGrid(tips_df, col='sex', hue='smoker', height=4, aspect=1.3,
                        hue_order=['Yes', 'No'],
												hue_kws=dict(marker=['^', 'v']))
tips_fg.map(sns.scatterplot, 'total_bill', 'tip', **kws)
```

```python
att_df = sns.load_dataset('attention')
att_fg = sns.FacetGrid(att_df, col='subject', col_wrap=5, height=1.5)
att_fg.map(sns.lineplot, 'solutions', 'score', marker='o')
```

```python
plt.figure(figsize=(8, 6))
sns.set_context('paper', font_scale=1.4)
sns.lmplot(x='total_bill', y='tip', hue='sex', data=tips_df, markers=['o', '^'], scatter_kws=dict(s=100, lw=0.5, edgecolor='w'))
```

```python
plt.figure(figsize=(8, 6))
sns.set_context('paper', font_scale=1.4)
sns.lmplot(
    x='total_bill', 
		y='tip', 
		hue='sex', 
		data=tips_df, 
		markers=['o', '^'], 
		scatter_kws=dict(s=100, lw=0.5, edgecolor='w'),
		ci=False
)
```

```python
sns.lmplot(x='total_bill', y='tip', col='sex', row='time', data=tips_df,
           height=8, aspect=0.6)
```
